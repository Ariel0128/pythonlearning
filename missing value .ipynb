```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
np.random.seed(25)
df = DataFrame(np.random.rand(36).reshape((6,6)))
df
```

```python
df.loc[3:5,0] = np.nan
```

```python
df.loc[1:4,5] = np.nan
df
```

### Fill missing values in different columns; if fill them with one value, just .fillna(9)

```python
df = df.fillna({0: 0.3, 5:0.5})
df
```

### Fill missing value with the last non-missing value in the column: 名字取的有问题，凑合看吧，会覆盖缺数据的df


```python
df = df.fillna(method='ffill')
df
```

### count the missing value

```python
df.isnull().sum()
```

### Remove the missing value

```python
df1 = df.dropna()
df1
```

```python
# remove the columns with missing value
```

```python
df2 = df.dropna(axis=1)
df2
```

```python
cls = pd.DataFrame(np.arange(6).reshape((6,1)))

df_class = pd.concat([df, cls], axis=1, ignore_index=True)
df_class.iloc[2:6,6] = 0
df_class
```

```python
hhh = df_class.groupby(df_class[6])
hhh.mean()
```

### replace missing value with the global mean

```python
for i in df_class:
    df_class[i]= df_class[i].fillna(df_class[i].mean())
```

```python
df_class
```

### replace missing value with the mean of the group

```python
df2 = pd.concat([df, cls], axis=1, ignore_index=True)
df2.iloc[2:6,6] = 0
df2
```

```python
df2[0].fillna(df2.groupby(6)[0].transform("mean"),inplace=True)
df2
```

```python
df2.groupby(6).mean()
```
