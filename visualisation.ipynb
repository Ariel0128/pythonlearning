### Object-oriented method

```python
%matplotlib inline
rcParams['figure.figsize']= 5,4        # the figures below will be originally 5 inches wide and 4 inches tall
```

```python
x = range(0, 4)
y = random(4)
tl = ['apple','orange','banana','kiwi']  # tl: tick labels

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])  # to control the unit of axes; trying different values will see

ax.set_ylim([0,1]) # if only set one axis (x or y), it will set two axes the same range.
ax.set_xlim([1,3])

ax.grid()

ax.set_xticks(range(4))  # tick marks  and will add grid automatically
ax.set_xticklabels(tl, rotation=60, fontsize='medium') # default rotation: horizontal no rotation; 

ax.set_title('the Name of this Plot')
ax.set_xlabel('number of me')

ax.legend('hahahaahahah', loc='upper left') # have no idea why it can't be shown

ax.annotate('this is the peak', xy= (3,0.8), xytext=(2,0.7), arrowprops=dict(facecolor='black', shrink=0.05))
 
# cause the y is randomised, thus it changes everytime... just see how it works.

ax.plot(x, y)
```
